```python
from covid19dh import covid19
import pandas as pd
```

```python
x2, _ = covid19("USA", level=2, verbose=False)
#x3, _ = covid19("USA", level=3, verbose=False)
```

```python
lean=pd.read_csv("../data/pol_lean.csv")
state_party_affiliation_2020_2021=pd.read_csv("../data/state_party_affiliation_2020_2021.csv")
```

```python
joint=x2.merge(state_party_affiliation_2020_2021,left_on="administrative_area_level_2", right_on="State")
joint=joint.merge(lean,left_on="administrative_area_level_2", right_on="State")
```

```python
joint.to_csv("../data/covid19dh_usa_party_joint.csv")
```
